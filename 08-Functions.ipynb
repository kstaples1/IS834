# Python Functions/Modules

In this quick section, we will run through functions, or modules.  I tend to call these functions, but you may see them referenced as modules elsewhere.

We have been exposed to functions we use object.`<tab>` to see the methods available to a given object.  At a high level, we can think of these modules as functions being applied to the object.

We `def`ine a function using the `def:`

Let's see some basic examples

## Quick example



```
# define a function to add two numbers
def add_it(x, y):
  types = (int, float)
  if type(x) not in types and type(y) not in types:
    print("you need to input x and y as int or float objects")
    return None
  return(x+y)
```

> Note:  There are better ways to handle this, but it helps keep things on track for now.

```
x = "a"
b = 1.0
z = add_it(x, b)
```

```
type(z)
```

```
add_it(1, 5.6)
```

```
# a function to make everything lowercase
def make_decision(value, criteria, greater=True):
  # if the value is greater, print a statement
  if greater and value > criteria:
    print("value is greater than criteria")
  elif greater == False and value < criteria:
    print("Value is less than critera")
  else:
    print("you conditions do not align")
```

```
make_decision(10, 2)
```

```
make_decision(10, 2, greater=False)
```

```
# replicate the length
def str_length(string):
  i = 0
  for s in string:
    i += 1
  return("the string has a length of {}".format(i))
```

```
str_length("Brock")
```

```
x = "Brock"
len(x)
```
