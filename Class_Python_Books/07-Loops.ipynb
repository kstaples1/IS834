# While Loops

While loops while continue to execute code until the condition is no longer **True**

````
# define the variables
i = 0


# loop check and print
while i < 5:
  print("{} is less than 5.  Adding 1 to i.".format(i))
  # incrementing i by 1
  i = i + 1  
````

We can also run code indefinately, until something changes.  

````
# set some variables
x = True
i = 1

while x:
  # divide i by 9 and get the remainder in the object check
  check = i % 9
  
  # check to see if divisible by 9
  if check == 0:
    print("i was divisible by 9. Stopping the while loop")
    x = False
  else:
    # print out the status
    print("x is still true.  the remainder was {}".format(check))
  
    # incrementing i.  Note a new way to do it!
    i += 1
  
  
````

There is another way to break the loop.  With `break`



---



## Break and Continue

````
# set some variables
x = True
i = 1

while x:
  # divide i by 9 and get the remainder in the object check
  check = i % 9
  
  # check to see if divisible by 9
  if check == 0:
    print("i was divisible by 9. Stopping the while loop")
    break
  else:
    # print out the status
    print("x is still true.  the remainder was {}".format(check))
  
    # incrementing i.  Note a new way to do it!
    i += 1
````

There are other options.  

> `continue` will stop the current iteration of the loop and kick off the next one, whereas break stops the loop all together.



````
i = 0
while i < 6:
  i += 1 
  if i == 3:
    continue
  print(i)
````

# For loops

For loops let us iterate over _something_ and perform actions, and optionally, apply conditions and break/continue.  Unlike the while loop, the `for` loop has a given number of "items" to consider.

````
# basic for loop
mystring = "Brock"

for s in mystring:
  print(s)
````

````
# edit a list to be double
mylist = [1,2,3,4,5]

## a new list
newlist = []

## loop and modify
for i in mylist:
  x = mylist[i-1]  ##<----- Note
  y = x * 2
  newlist.append(y)
````

````
print(mylist)
print(newlist)
````
